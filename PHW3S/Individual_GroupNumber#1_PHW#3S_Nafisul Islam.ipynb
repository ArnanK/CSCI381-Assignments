## Proposition 1
This SQL query retrieves customers from **both Territory 1 and Territory 2**, displaying their **CustomerID**, **Customer Name**, and **Territory**. It uses the **`UNION`** operation to combine the results from the two territories into a single list.

### Steps:

1. **First Query (Customers in Territory 1)**:
    - Retrieves the **CustomerID**, **Customer Name**, and **Territory Name** for customers who placed orders in **Territory 1**.
    - Joins the **`Sales.Customer`**, **`Person.Person`**, **`SalesOrderHeader`**, and **`SalesTerritory`** tables to link customer data with their associated territory.
    - Filters for customers in **Territory 1** using **`st.TerritoryID = 1`**.

2. **Second Query (Customers in Territory 2)**:
    - Retrieves the same details (i.e., **CustomerID**, **Customer Name**, and **Territory Name**) for customers who placed orders in **Territory 2**.
    - Filters for customers in **Territory 2** using **`st.TerritoryID = 2`**.

3. **`UNION`**:
    - The **`UNION`** operation combines the results from both queries, returning a unified list of customers from **both territories**.
    - **`UNION`** removes duplicate rows (if any), so a customer who placed orders in both territories would only appear once.

### Output:
- The result will display **CustomerID**, **Customer Name**, and **Territory** for customers who placed orders in **either Territory 1 or Territory 2**.

```sql
USE Adventureworks2019;
GO

SELECT 
    c.CustomerID,
    p.FirstName + ' ' + p.LastName AS CustomerName,
    st.Name AS Territory
FROM 
    Sales.Customer AS c
JOIN 
    Person.Person AS p ON c.PersonID = p.BusinessEntityID
JOIN 
    Sales.SalesOrderHeader AS soh ON c.CustomerID = soh.CustomerID
JOIN 
    Sales.SalesTerritory AS st ON soh.TerritoryID = st.TerritoryID
WHERE 
    st.TerritoryID = 1

UNION

SELECT 
    c.CustomerID,
    p.FirstName + ' ' + p.LastName AS CustomerName,
    st.Name AS Territory
FROM 
    Sales.Customer AS c
JOIN 
    Person.Person AS p ON c.PersonID = p.BusinessEntityID
JOIN 
    Sales.SalesOrderHeader AS soh ON c.CustomerID = soh.CustomerID
JOIN 
    Sales.SalesTerritory AS st ON soh.TerritoryID = st.TerritoryID
WHERE 
    st.TerritoryID = 2;

```

This SQL query retrieves the **Business Entity ID**, the full name of each **Salesperson**, and their **Average Sales**. The average sales are calculated based on the unit price multiplied by the order quantity for all orders handled by the salesperson.

- **`AVG(sod.UnitPrice * sod.OrderQty)`**: Calculates the average sales amount for each salesperson.
- **`JOIN`**: Multiple joins are used to link sales order details with sales order headers, salespersons, and their personal details.
- **`GROUP BY sp.BusinessEntityID, p.FirstName, p.LastName`**: Groups the results by salesperson, ensuring each salesperson's average sales are calculated.
- **`ORDER BY AverageSales DESC`**: Orders the result in descending order to show salespersons with the highest average sales at the top.

```sql

USE AdventureWorks2019;
GO

SELECT 
    sp.BusinessEntityID,
    p.FirstName + ' ' + p.LastName AS Salesperson,
    AVG(sod.UnitPrice * sod.OrderQty) AS AverageSales
FROM 
    Sales.SalesOrderDetail sod
JOIN 
    Sales.SalesOrderHeader soh ON sod.SalesOrderID = soh.SalesOrderID
JOIN 
    Sales.SalesPerson sp ON soh.SalesPersonID = sp.BusinessEntityID
JOIN 
    Person.Person p ON sp.BusinessEntityID = p.BusinessEntityID
GROUP BY 
    sp.BusinessEntityID, p.FirstName, p.LastName
ORDER BY 
    AverageSales DESC;
```

## Proposition 7

This SQL query calculates the **Total Profit** for each product sold between **July 1, 2011**, and **August 1, 2011**. The total profit for each product is calculated by subtracting the **Standard Cost** from the **Unit Price**, multiplied by the number of units sold (OrderQty), and then summing this value for all sales.

- **`SUM((sod.UnitPrice * sod.OrderQty) - (p.StandardCost * sod.OrderQty)) AS TotalProfit`**: This calculates the total profit for each product by multiplying the unit price and order quantity, subtracting the product of the standard cost and order quantity, and summing the result.
- **`LEFT JOIN`**: The `LEFT JOIN` ensures that all products are included in the result set, even if they do not have corresponding sales orders.
- **`WHERE soh.OrderDate BETWEEN '2011-07-01' AND '2011-08-01'`**: Filters the results to include only sales orders placed within the specified date range.
- **`GROUP BY p.ProductID, p.Name`**: Groups the results by **ProductID** and **Product Name** so that each product's total profit is calculated individually.

```sql
USE AdventureWorks2019;
GO

SELECT 
    p.ProductID,
    p.Name,
    SUM((sod.UnitPrice * sod.OrderQty) - (p.StandardCost * sod.OrderQty)) AS TotalProfit
FROM 
    Sales.SalesOrderDetail sod
LEFT JOIN 
    Production.Product p ON sod.ProductID = p.ProductID
LEFT JOIN 
    Sales.SalesOrderHeader soh ON soh.SalesOrderID = sod.SalesOrderID
WHERE 
    soh.OrderDate BETWEEN '2011-07-01' AND '2011-08-01'
GROUP BY 
    p.ProductID, p.Name;
```

## proposition 8

This SQL query identifies the most popular credit card types used in each sales **territory**. The query first calculates the number of customers using each credit card type by territory and then identifies the most used card types in each region. It uses a **Common Table Expression (CTE)** to simplify the query structure.

### Steps:
1. **CTE (`CreditCardUsageByTerritory`)**:
    - This Common Table Expression (CTE) calculates the number of customers using each credit card type in each sales territory.
    - It joins the **CreditCard**, **PersonCreditCard**, **SalesOrderHeader**, and **SalesTerritory** tables to connect customer credit card usage with their respective territories.
    - The result is grouped by **Territory Name** and **Credit Card Type**, and **`COUNT(DISTINCT pcc.BusinessEntityID)`** is used to ensure unique customer counts.

2. **Main Query**:
    - The main query selects the **Territory Name**, **Credit Card Type**, and the **MostUsed** credit card type (the maximum number of customers using that card type in a territory).
    - **`MAX(NumberOfCustomers)`** ensures that the most used credit card type for each territory is identified.
    - The results are grouped by **Territory Name** and **Credit Card Type**, and ordered first by **Territory Name** and then by the highest customer count (**MostUsed**) in descending order.

### Output:
- The result shows the **Territory Name**, **Credit Card Type**, and the **MostUsed** count for each territory, providing insights into which credit card types are most popular in different regions.

```sql
USE AdventureWorks2019;
GO

WITH CreditCardUsageByTerritory AS (
    SELECT 
        st.Name AS TerritoryName,
        cc.CardType AS CreditCardType,
        COUNT(DISTINCT pcc.BusinessEntityID) AS NumberOfCustomers
    FROM 
        Sales.CreditCard AS cc
    JOIN 
        Sales.PersonCreditCard AS pcc ON cc.CreditCardID = pcc.CreditCardID
    JOIN 
        Sales.SalesOrderHeader AS soh ON soh.CustomerID = pcc.BusinessEntityID
    JOIN 
        Sales.SalesTerritory AS st ON soh.TerritoryID = st.TerritoryID
    GROUP BY 
        st.Name, cc.CardType
)

SELECT 
    TerritoryName,
    CreditCardType,
    MAX(NumberOfCustomers) AS MostUsed
FROM 
    CreditCardUsageByTerritory
GROUP BY 
    TerritoryName, CreditCardType
ORDER BY 
    TerritoryName, MostUsed DESC;

```

This SQL query retrieves a list of **products** that either contain the term **'Gloves'** or **'Jersey'** in their name. It uses the **`UNION ALL`** set operation to combine the results from two different queries without removing duplicates.

### Steps:
1. **First Query (Products with 'Gloves' in Name)**:
   - This part retrieves the **Product Name** and **Product Color** for all products whose name includes the term **'Gloves'**. 
   - The **`LIKE '%Gloves%'`** condition filters for products containing the word "Gloves" in their name.

2. **Second Query (Products with 'Jersey' in Name)**:
   - Similarly, this part retrieves the **Product Name** and **Product Color** for all products whose name includes the term **'Jersey'**. 
   - The **`LIKE '%Jersey%'`** condition filters for products containing the word "Jersey" in their name.

3. **`UNION ALL`**:
   - The **`UNION ALL`** operation combines the results from both queries. Unlike **`UNION`**, which removes duplicates, **`UNION ALL`** keeps all entries, even if they are duplicated between the two queries.
   
### Output:
- The result will display a list of products that have either **'Gloves'** or **'Jersey'** in their name, along with their respective colors.

```sql
USE AdventureWorks2019;
GO


SELECT 
    p.Name AS ProductName, 
    p.Color AS ProductColor
FROM 
    Production.Product AS p
WHERE 
    p.Name LIKE '%Gloves%'

UNION ALL

SELECT 
    p.Name AS ProductName, 
    p.Color AS ProductColor
FROM 
    Production.Product AS p
WHERE 
    p.Name LIKE '%Jersey%';


```
