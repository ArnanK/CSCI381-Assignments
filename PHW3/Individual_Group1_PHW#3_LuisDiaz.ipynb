```sql
/*
Proposition 1. List all customers and all suppliers in a cross join.
This query list of every possible combination of customers and suppliers using a CROSS JOIN
*/
Select C.[CustomerID], C.[CustomerName], S.[SupplierID], S.[SupplierName]
From [Sales].[Customers] AS C
Cross join [Purchasing].[Suppliers] AS S

/*
Proposition 2. Find all purchase orders placed between the years 2012 and 2016.
filters purchase orders placed between 2012 and 2016 using the YEAR() function.
*/
select [PurchaseOrderID], [SupplierID], [OrderDate], [ExpectedDeliveryDate]
from [Purchasing].[PurchaseOrders]
where year([OrderDate]) Between 2012 and 2016

/*
Proposition 3. Retrieve details of all invoices issued between the years 2013 and 2015 using a BETWEEN filter.
*/
select [InvoiceID], [CustomerID], [InvoiceDate], [TotalDryItems], [TotalChillerItems]
from [Sales].[Invoices]
where year([InvoiceDate]) Between 2013 AND 2015

/*
Proposition 4. List all customers along with the buying group they belong to.
retrieves all customers and their associated buying group names, if any, using a LEFT JOIN to include customers without a buying group.
*/
Select C.[CustomerID], C.[CustomerName], BG.[BuyingGroupName]
from [Sales].[Customers] AS C
left join [Sales].[BuyingGroups] AS BG ON C.[BuyingGroupID] = BG.[BuyingGroupID]

/*
Proposition 5. List all suppliers and their primary contact details.
This query retrieves all suppliers along with their primary contact details using a LEFT JOIN to include suppliers without contact information.
*/
Select S.[SupplierID], S.[SupplierName], P.[FullName] AS [ContactName], P.[PhoneNumber] AS [ContactPhone]
From [Purchasing].[Suppliers] AS S
Left join [Application].[People] AS P ON S.[PrimaryContactPersonID] = P.[PersonID]

/*
Proposition 6. This query calculates the average temperature for all warehouse vehicles
using a WHERE clause to filter records between 2010 and 2016.
*/
Select AVG([Temperature]) AS [AverageVehicleTemperature]
from [Warehouse].[VehicleTemperatures]
Where YEAR([RecordedWhen]) BETWEEN 2010 AND 2016

/*
Proposition 7. Count the number of suppliers by each supplier category.
This query uses an INNER JOIN to link suppliers to their categories, and a GROUP BY clause
*/
Select SC.[SupplierCategoryName], COUNT(S.[SupplierID]) AS [SupplierCount]
From [Purchasing].[Suppliers] AS S
Inner join [Purchasing].[SupplierCategories] AS SC ON S.[SupplierCategoryID] = SC.[SupplierCategoryID]
Group by SC.[SupplierCategoryName]

/*
8. Show all stock items that have not been sold (i.e., have no related orders).
This query retrieves all stock items that have never been sold using a LEFT JOIN to link stock items to orders
*/
SELECT SI.[StockItemID], SI.[StockItemName]
FROM [Warehouse].[StockItems] AS SI
LEFT JOIN [Sales].[OrderLines] AS OL ON SI.[StockItemID] = OL.[StockItemID]


/*
Proposition 9. Get the number of purchase orders for each supplier placed after 2011.
This query filters purchase orders placed after 2011 and uses a GROUP BY clause to count orders for each supplier.
*/
SELECT [SupplierID], COUNT([PurchaseOrderID]) AS [PurchaseOrderCount]
FROM [Purchasing].[PurchaseOrders]
WHERE YEAR([OrderDate]) > 2011
GROUP BY [SupplierID]

/*
Proposition 10. List customers who have made transactions and the total amount of their transactions.
This query uses an INNER JOIN to link customers to their transactions
*/
SELECT C.[CustomerID], C.[CustomerName], SUM(CT.[AmountExcludingTax]) AS [TotalTransactions]
FROM [Sales].[Customers] AS C
INNER JOIN [Sales].[CustomerTransactions] AS CT ON C.[CustomerID] = CT.[CustomerID]
GROUP BY C.[CustomerID], C.[CustomerName]
```
