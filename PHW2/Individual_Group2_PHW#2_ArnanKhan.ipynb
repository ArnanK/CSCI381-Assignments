**<u>Propoistion 1:</u>** Find the average annual salary for the states in the US within the database (Highest-Lowest).  

_(Note: Not all states will display as the employees/persons in the database do not work/live in particular states.)_

```sql
WITH PersonUS_CTE AS
(
SELECT P.BusinessEntityID, SP.[Name] as [State]
FROM Person.Person as P
INNER JOIN Person.BusinessEntity as BE ON BE.BusinessEntityID = P.BusinessEntityID
INNER JOIN Person.BusinessEntityAddress as BEA ON BE.BusinessEntityID = BEA.BusinessEntityID
INNER JOIN Person.Address as A ON A.AddressID = BEA.AddressID
INNER JOIN Person.StateProvince as SP ON A.StateProvinceID = SP.StateProvinceID
WHERE SP.CountryRegionCode = 'US'
)

--Rate is hourly. 35 work hours a week and 52 weeks a year.
SELECT P.[State],  CONCAT('$',AVG(EPH.Rate * 35 * 52) ) AS [Average Annual Salary]
FROM PersonUS_CTE as P
INNER JOIN HumanResources.Employee AS E ON P.BusinessEntityID = E.BusinessEntityID
INNER JOIN HumanResources.EmployeePayHistory AS EPH ON EPH.BusinessEntityID = E.BusinessEntityID
GROUP BY P.[State]
ORDER BY [Average Annual Salary] DESC
```

**<u>Proposition 2:</u>** What is the headcount of all active employees by each department and annual costs for pay.

```sql
SELECT D.Name, COUNT(E.BusinessEntityID) as [Head Count], CONCAT('$',SUM(EPH.Rate * 35 * 52)) as [Total Employee Pay]
FROM HumanResources.Employee as E
INNER JOIN HumanResources.EmployeeDepartmentHistory as EDH ON E.BusinessEntityID = EDH.BusinessEntityID
INNER JOIN HumanResources.Department as D ON EDH.DepartmentID = D.DepartmentID
INNER JOIN HumanResources.EmployeePayHistory As EPH ON EPH.BusinessEntityID = E.BusinessEntityID
--WHERE D.Name = 'Production'
GROUP BY D.Name
ORDER BY [Total Employee Pay] DESC
```

Proposition 3: Who are all the people in the database that are not aprat of HumanResouces.Employee

```sql
SELECT CONCAT(P.FirstName, ' ', P.LastName) as [Full Name], E.BusinessEntityID
FROM Person.Person as P
LEFT OUTER JOIN HumanResources.Employee as E ON E.BusinessEntityID = P.BusinessEntityID
WHERE E.BusinessEntityID is NULL
```
